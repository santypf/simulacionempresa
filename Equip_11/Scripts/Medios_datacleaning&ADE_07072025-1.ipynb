```python
import pandas as pd
import requests
from io import BytesIO

# URL directa del archivo Excel
url = "https://www.mites.gob.es/estadisticas/EAL/EAL2023/Tablas_EAL_2023.xlsx"

# Descargar archivo
response = requests.get(url)
excel_file = BytesIO(response.content)

# Leer la hoja 'EAL-18a' desde la fila 7 (índice 6)
df = pd.read_excel(
    excel_file,
    sheet_name="EAL-18a",
    engine="openpyxl",
    skiprows=6  # Empieza en la fila 7 de Excel
)

```

```python
df
```

```python
df = df.rename(columns={"Unnamed: 0": "Categoría"})

filtros = [
    "TOTAL",
    "TAMAÑO DE LA EMPRESA",
    "ACTIVIDAD ECONÓMICA",
    "COMUNIDAD AUTÓNOMA"
]
df = df[~df["Categoría"].isin(filtros)]
```

```python
df = df.reset_index(drop=True)
df = df.dropna(how="any")

df
```

```python
df = df.rename(columns={"Unnamed: 0": "Categoría"})

# Define los grupos manualmente (según tus datos)
empresa = [
    "De 5 a 9 trabajadores", 
    "De 10 a 49 trabajadores", 
    "De 50 a 249 trabajadores", 
    "De 250 a 499 trabajadores", 
    "Más de 499 trabajadores"
]

actividad = [
    "Industria", "Construcción", "Comercio y reparación de vehículos",
    "Transporte y almacenamiento", "Hostelería",
    "Información, comunicaciones, actividades financieras, de seguros e inmobiliarias",
    "Actividades profesionales, científicas y técnicas",
    "Actividades administrativas y servicios auxiliares",
    "Actividades artísticas, recreativas y de entretenimiento y otros servicios"
]

region = [
    "Andalucía", "Aragón", "Asturias (Principado de)", "Balears (Illes)", "Canarias",
    "Cantabria", "Castilla-La Mancha", "Castilla y León", "Cataluña", "Comunitat Valenciana",
    "Extremadura", "Galicia", "Madrid (Comunidad de)", "Murcia (Región de)",
    "Navarra (Comunidad Foral de)", "País Vasco", "Rioja, La"
]
```

```python
def clasificar_categoria(categoria):
    cat = str(categoria).strip().lower()
    if cat in [x.lower() for x in actividad]:
        return "Actividad económica"
    elif cat in [x.lower() for x in region]:
        return "Comunidad autónoma"
    elif cat in [x.lower() for x in empresa]:
        return "Tamaño de empresa"
    else:
        return "Otro"

df["Tipo"] = df["Categoría"].apply(clasificar_categoria)
df
```
